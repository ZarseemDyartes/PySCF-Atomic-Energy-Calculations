```python
from pyscf import gto,scf,cc
import basis_set_exchange as bse
import matplotlib.pyplot as plt
element = [['In',1,bse.get_basis('STO-2G',elements=[49],fmt='nwchem',header=False)],['Sn',2,bse.get_basis('STO-2G',elements=[50],fmt='nwchem',header=False)],['Sb',3,bse.get_basis('STO-2G',elements=[51],fmt='nwchem',header=False)],['Te',2,bse.get_basis('STO-2G',elements=[52],fmt='nwchem',header=False)],['I',1,bse.get_basis('STO-2G',elements=[53],fmt='nwchem',header=False)],['Xe',0,bse.get_basis('STO-2G',elements=[54],fmt='nwchem',header=False)]]
rhf = []
data = [['STO-2G']]
for dp in range(len(element)):
    rhf.append([])
    
def hf(a,b,s,n):
    mol = gto.Mole()
    mol.build(
        symmetry = 'D2h',
        atom = [[a,(0,0,0)], ],
            #a - String variable, input symbol of element
        basis = gto.basis.parse(b),
            #bset - String variable, input basis set
        spin = s
            #s - Integer variable, input number of unpaired electrons
    )
    mol.verbose = 0
    mf = mol.RHF().run()
    rhf[n].append(mf.e_tot)
            #n - Integer variable
            
for atom in element:
    hf(atom[0],atom[2], atom[1],element.index(atom))
data[0].insert(0,'RHF Energy')
for energy in range(len(rhf)):
    data.append(rhf[energy])
for symbol in range(len(element)):
    rhf[symbol].insert(0,element[symbol][0])
    
fig, ax = plt.subplots()
table2 = ax.table(cellText=data, loc = 'center')
table2.set_fontsize(25)
table2.scale(10,5)
ax.axis('off')
plt.show()
```
