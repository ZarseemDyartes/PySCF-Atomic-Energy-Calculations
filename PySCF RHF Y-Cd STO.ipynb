```python
from pyscf import gto,scf,cc
import basis_set_exchange as bse
import matplotlib.pyplot as plt
element = [['Y',1,bse.get_basis('STO-2G',elements=[39],fmt='nwchem',header=False)],['Zr',2,bse.get_basis('STO-2G',elements=[40],fmt='nwchem',header=False)],['Nb',3,bse.get_basis('STO-2G',elements=[41],fmt='nwchem',header=False)],['Mo',6,bse.get_basis('STO-2G',elements=[42],fmt='nwchem',header=False)],['Tc',5,bse.get_basis('STO-2G',elements=[43],fmt='nwchem',header=False)],['Ru',4,bse.get_basis('STO-2G',elements=[44],fmt='nwchem',header=False)],['Rh',3,bse.get_basis('STO-2G',elements=[45],fmt='nwchem',header=False)],['Pd',2,bse.get_basis('STO-2G',elements=[46],fmt='nwchem',header=False)],['Ag',1,bse.get_basis('STO-2G',elements=[47],fmt='nwchem',header=False)],['Cd',0,bse.get_basis('STO-2G',elements=[48],fmt='nwchem',header=False)]]
rhf = []
data = [['STO-2G']]
for dp in range(len(element)):
    rhf.append([])
    
def hf(a,b,s,n):
    mol = gto.Mole()
    mol.build(
        symmetry = 'D2h',
        atom = [[a,(0,0,0)], ],
            #a - String variable, input symbol of element
        basis = gto.basis.parse(b),
            #bset - String variable, input basis set
        spin = s
            #s - Integer variable, input number of unpaired electrons
    )
    mol.verbose = 0
    mf = mol.RHF().run()
    rhf[n].append(mf.e_tot)
            #n - Integer variable
            
for atom in element:
    hf(atom[0],atom[2], atom[1],element.index(atom))
data[0].insert(0,'RHF Energy')
for energy in range(len(rhf)):
    data.append(rhf[energy])
for symbol in range(len(element)):
    rhf[symbol].insert(0,element[symbol][0])
    
fig, ax = plt.subplots()
table2 = ax.table(cellText=data, loc = 'center')
table2.set_fontsize(25)
table2.scale(10,5)
ax.axis('off')
plt.show()
```
